# Shark Attack Data Analysis

# Abstract
Geert Hofstede has developed an index of cultural insights which demonstrate a country's varying levels of indulgence, individualism, uncertainty avoidance, risk aversion, etc. One might expect a country with a high level of risk aversion to engage less in dangerous activities like swimming in shark-infested waters. This paper will prove or disprove whether there is a correlation between the two data sets. Shark attack data in this paper comes from Kaggle, a free online resource, and has poor readability – it is difficult data to work with. The Kaggle data provided includes over 6,300 unique shark attacks ranging from eater sports to accidents. Geert Hofstede is a well-funded company and so their data is easy to work with.

I am losing valuable data for some low-count countries like Mexico (rank 8, 89 values) by 
removing countries with values under 10, so I need to use regex to merge countries with 
names similar to each other. This includes:

- Strip titles
- Set titles to lowercase
- Remove special chars ()[]?
- Remove /, ocean, and sea values

```python
# Change all values to lowercase
data['Country'] = data['Country'].apply(str.lower)

# strip values
data['Country'] = data['Country'].apply(lambda x: x.strip())

# Remove special characters from value titles: ,?()[]
# data.strings.str.replace('[^a-z]', '')
data['Country'] = data['Country'].apply(lambda x: re.sub(r'(\.|\,|\(|\))', '', x))

# Remove rows with titles including /, ocean, or sea just to visualize data better (since these values do not count as unique countries)
data['Country'] = data['Country'].apply(rmv_ocean)

# redisplay data
print('unique country values: ', len(pd.unique(data['Country'])), '\n', pd.unique(data['Country']))
```

## 5. Import relevant data to compare for hypothesis testing.
- Import country's 'independence levels' (from Geert Hofstede Insights) into a dictionary.
- Convert that to a dataframe.
- Add country population size.
- convert shark attack data into new column reflecting attack data 'per capita.'

```python
# Importing Geert Hofstede's country data manually, as there is no PDF file for this interactive data
independence_score = {'usa':{'indulgence':68, 'individualism':91, 'uncertainty':46}, 'australia':{'indulgence':71, 'individualism':90, 'uncertainty':51}, 
                      'south africa':{'indulgence':63, 'individualism':65, 'uncertainty':49}, 'new zealand':{'indulgence':75, 'individualism':79, 'uncertainty':49}, 
                      'brazil':{'indulgence':59, 'individualism':38, 'uncertainty':76}, 'mexico':{'indulgence':97, 'individualism':30, 'uncertainty':82}, 
                      'italy':{'indulgence':30, 'individualism':76, 'uncertainty':75}, 'fiji':{'indulgence':0, 'individualism':14, 'uncertainty':48}, 
                      'philippines':{'indulgence':42, 'individualism':32, 'uncertainty':44}, 'mozambique':{'indulgence':80, 'individualism':15, 'uncertainty':44}, 
                      'spain':{'indulgence':44, 'individualism':51, 'uncertainty':86}, 'india':{'indulgence':26, 'individualism':48, 'uncertainty':40}, 
                      'egypt':{'indulgence':0, 'individualism':37, 'uncertainty':55}, 'croatia':{'indulgence':33, 'individualism':33, 'uncertainty':80}, 
                      'japan':{'indulgence':42, 'individualism':46, 'uncertainty':92}, 'panama':{'indulgence':0, 'individualism':11, 'uncertainty':86}
                     }
i_score = pd.DataFrame(data=independence_score).T

# save country_data rows that are present in i_score to a new dataframe
i_score['shark_attacks'] = country_data['Country'].value_counts()

# Import country population size, per 1,000
i_score.loc[:,'population'] = [329500, 25690, 59310, 5084, 212600, 128900, 59550, 896, 109600, 31260, 47350, 1380000, 102300, 4047, 125800, 4315]

# Calculate shark attack 'per capita' rate
i_score = i_score.assign(per_capita=lambda x: x.shark_attacks/x.population)

i_score
```

# HYPOTHESIS DISPROVEN (with evidence)
There is not a correlation between a country's individualism or even uncertainty avoidance and the number of shark attacks in the area.

### A correlation of .67 is high, however...
the data compared here is not adjusted per capita and therefore cannot be used to compare large countries to small countries on amount of shark attacks.

### The individualism and indlugence charts...
display a clearly random pattern, demosntrating little to no correlation between cultural values and shark attacks in a given country. The correlation between the data before the 'per capita' adjustment can almost be seen because America has the highest of these scores with the most shark attacks, but South Africa skews the data with a lower score and a relatively large number of attacks.

### In the future...
it would be beneficial to gather data on foreigner/domestic attack victims to determine whether tourist-reliant countries like Fiji really do encounter so many shark attacks because of locals or people of other cultures.
